# Simple Sampling Experiments

# 2.1 Sequences of Fair Coins Flips Can Be Understood as sampling the Uniform distribution on [0,1]

Using coin flips we can map a sequence of $p$ coins flips onto a number between 0 and 1. We thereby use the following mapping rule:
\begin{equation}
    n = \sum_{j=1}^p \frac{f_j}{2^j}
\end{equation}
where $f_j$ is equal to 1 of the coin lands heads and 0 if it lands tails.

 ## Minimum value for n:

Using $p = 16$ we can calculate the minimum possible value for n. If all 16 coin flip land tails we obtain:
\begin{equation}
    n = \sum_{j=1}^{16} \frac{f_j}{2^j} = \sum_{j=1}^{16} \frac{0}{2^j} = 0
\end{equation}

## Maximum value for n:

Using $p = 16$ we can calculate the maximum possible value for n. If all 16 coin flip land heads we obtain:
\begin{equation}
    n = \sum_{j=1}^{16} \frac{f_j}{2^j} = \sum_{j=1}^{16} \frac{1}{2^j} = 0.99987
\end{equation}

## Maximum value for n with $p \rightarrow \infty$
Using $p \rightarrow \infty$ we can calculate the maximum possible value for n. If all coin flip land heads we obtain:
\begin{equation}
    n = \sum_{j=1}^\infty \frac{f_j}{2^j} = \sum_{j=1}^\infty \frac{1}{2^j} = \sum_{j=0}^\infty \Bigl (\frac{1}{2}\Bigr)^j -1 = \frac{1}{1-\frac{1}{2}} -1 = 2-1 = 1  
\end{equation}
where we made use of the geometric series.

```python
p=32
for S in [10**2,10**4,10**6]:
    real = np.zeros(S)
    sequences = np.random.randint(2, size=(p, S)) #Flips coins
    for sequence in range(S):
        real_sequence = 0
        for j in range(1,p+1):
            real_sequence += sequences[j-1,sequence]/2**j
        real[sequence] = real_sequence
    plt.hist(real,bins=20)
    plt.xlabel("Real numbers")
    plt.ylabel("Frequency")
    plt.title(f"S = {S} sequences")
    plt.savefig(f"sequence_{S}.png")
    plt.show()
```

# 2.2 Independent Coins with Different Distributions

In this exercise we;;ll
see that if the different flips in a sequence have different distributions we get a different distribution for the real observable.
To achieve that, we will flip coins with different biases, so that the p flips are independent but not identically distributed.

Suppose the jth coin has $$P_{j}(\mathrm{tails})=1-P_{j}(\mathrm{heads})=\frac{1}{1+e^{-\lambda/2^{j}}}.$$

For $λ ∈ 0, 0.5, 1, 2$ we sample $10^6$ sequences of $p = 32$ different coins; the jth coin in each sequence should be biased as in the equation mentioned above. Then, we evaluate the real observable on each sequence, interpreting these flips as $10^6$ real observables in $[0, 1]$.

Finally, we make a histogram with our results for each $λ$ as it can be seen below:

P.S.: As the code below is 0-indexed, we replace $j$ with $j+1$.

```python
def calculate_p_tails(j, lambda_val):
    """Calculate P_j(tails) for given j and lambda."""
    return 1 / (1 + np.exp(-lambda_val/2**(j+1)))

def binary_sequence_to_real(sequence):
    """Convert binary sequence to real number in [0,1]."""
    return sum(bit * 2**(-i-1) for i, bit in enumerate(sequence))

def run_simulation(lambda_val, p=32, n_samples=1000000):
    """Run simulation for given lambda value."""
    # Pre-calculate probabilities for each position
    p_tails = np.array([calculate_p_tails(j, lambda_val) for j in range(p)])

    # Generate random numbers for all samples at once
    random_numbers = np.random.random((n_samples, p))

    # Generate coin flips (1 for tails, 0 for heads)
    sequences = (random_numbers < p_tails).astype(int)

    # Convert sequences to real numbers
    powers = 2.0 ** (-np.arange(1, p+1))
    real_observables = sequences @ powers

    return real_observables

# Set up the plot
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()
lambda_values = [0, 0.5, 1, 2]

for idx, lambda_val in enumerate(lambda_values):
    # Run simulation
    real_observables = 1-run_simulation(lambda_val)

    # Create histogram
    axes[idx].hist(real_observables, bins=100, density=True, alpha=0.7)
    axes[idx].set_title(f'λ = {lambda_val}')
    axes[idx].set_xlabel('Real Observable')
    axes[idx].set_ylabel('Density')

plt.tight_layout()
plt.show()

# Print some summary statistics
for lambda_val in lambda_values:
    real_observables = run_simulation(lambda_val)
    print(f"\nStatistics for λ = {lambda_val}:")
    print(f"Mean: {np.mean(real_observables):.4f}")
    print(f"Std Dev: {np.std(real_observables):.4f}")
    print(f"Min: {np.min(real_observables):.4f}")
    print(f"Max: {np.max(real_observables):.4f}")
```

We can observe that, because the coins have biases, the induced probability distribution on the real observable is not flat.

The real observables x are distributed exponentially on the interval $[0, 1]$. The
probability density, up to normalization, can be written as:
$$P(x)\ \propto\,e^{-\,\lambda\,x}$$
So the probability density with the normalization constant can be written as:
$$P(x)\ =\cal{N}e^{-\,\lambda\,x}$$
In the following goal, we find the normalization constant of $P$ and we plot the normalized $P(x)$ on top of the respective histogram:

```python
def theoretical_density(x, lambda_val):
    """Calculate theoretical probability density."""
    global normalization
    if lambda_val == 0:  # Handle special case
        return np.ones_like(x)  # Uniform distribution
    else:
        normalization = lambda_val / (1 - np.exp(-lambda_val))
        return normalization * np.exp(-lambda_val * x)
# Set up the plot
fig, axes = plt.subplots(2, 2, figsize=(10, 8))
axes = axes.ravel()
lambda_values = [0, 0.5, 1, 2]
for idx, lambda_val in enumerate(lambda_values):
    # Run simulation
    real_observables = 1-run_simulation(lambda_val)
    # Create histogram
    axes[idx].hist(real_observables, bins=100, density=True, alpha=0.7,

                  label='Simulation', color='skyblue')
    # Add theoretical distribution
    x = np.linspace(0, 1, 1000)
    y = theoretical_density(x, lambda_val)
    axes[idx].plot(x, y, 'r-', label='Theory', linewidth=2)
    axes[idx].set_title(f'λ = {lambda_val}')
    axes[idx].set_xlabel('Real Observable')
    axes[idx].set_ylabel('Density')
    axes[idx].legend()

plt.tight_layout()
plt.show()
print(f"Normalization constant: {normalization:.4f}")
```

The final result for the normalization constant is $\cal{N}\approx$$2.313$

# 2.3 Transforming CDFs

The CDF of the standard normal distribution for $\mu=0$ and $\sigma=1$ $(Z\sim \cal{N}$$(0,1))$ is given by:
$$CDF_{\mu=0,\sigma=1}(z)=\frac{1}{2}\left(1+\exp\left(\frac{z}{\sqrt{2}}\right)\right)$$
For a normal distribution
$X\sim \cal{N}$$(\mu,\sigma^{2})$, we can standardize
$X$ to obtain a standard normal variable
$Z$ as follows:
$$Z={\frac{X-\mu}{\sigma}}$$
Thus, $X\leq x$ is equivalent to $Z\leq{\frac{x-\mu}{\sigma}}$.

The CDF of $X$ is the probability that $X\leq x$:
$$CDF_{\mu,\sigma}(x)=P(X\leq x)=P\left(Z\leq{\frac{x-\mu}{\sigma}}\right)$$
Now, we substitute the expression for
$CDF(z)$ in terms of the error function:
$$CDF_{\mu,\sigma}(x)=CDF\left({\frac{x-\mu}{\sigma}}\right)={\frac{1}{2}}\left(1+\operatorname{erf}\left({\frac{x-\mu}{\sigma{\sqrt{2}}}}\right)\right)$$

# 2.4 Inverse Transform sampling

```python
def CDF(x,μ,s):
    return ((1+np.tanh((x-μ)/(2*s)))/2)**2
```

Inverse Function:
\begin{align}
    u &= \Biggl(\frac{1+\tanh{\frac{x-\mu}{2s}}}{2}\Biggr)^2 \iff 2u^{\frac{1}{2}}-1 = \tanh{\frac{x-\mu}{2s}}\\
    & \iff x = \text{arctanh}\bigl({2 u^{\frac{1}{2}}-1}\bigr)2s+ \mu
\end{align}

```python
μ = 0
s=3

us = np.random.uniform(0,1,10**5)
xs = np.arctanh(2*np.sqrt(us)-1)*2*s+μ #inverse of CDF
```

# Test the inverse function:

```python
u = np.random.uniform(0,1,10**3) # Use less datapoints
x = np.arctanh(2*np.sqrt(u)-1)*2*s+μ
x_ind = np.linspace(-11,25,10000)
plt.xlim(-11,25)
plt.plot(x_ind,CDF(x_ind,μ,s), label="plotted with CDF",linewidth=4,alpha=0.7)
plt.scatter(x,u,label="plotted with inverse of CDF",color="red",s=4)
plt.legend()
plt.xlabel("x")
plt.ylabel("CDF")
plt.grid()
plt.show()
```

# Histogram of x

```python
plt.hist(xs,bins=50)
plt.xlabel("x")
plt.ylabel("Frequency")
plt.vlines(np.mean(xs),0,13000,color="red",label=f"Mean m = {round(np.mean(xs),2)}")
plt.ylim(0,13000)
plt.title("Inverse Transform Sampling")
plt.legend()
plt.show()
```

# Analytical PDF

We want to calculate the PDF analytcally from the CDF:
\begin{align}
    &PDF(x) = \frac{\partial CDF(x)}{\partial x} = 2\Biggl(\frac{1+\tanh{\frac{x-\mu}{2s}}}{2}\Biggr) \cdot \frac{1}{2} \cdot \frac{1}{\cosh{\frac{x-\mu}{2s}}^2} \cdot \frac{1}{2s} \\
    & = \frac{1}{4s} \frac{1+\tanh{\frac{x-\mu}{2s}}}{\cosh{\frac{x-\mu}{2s}}^2} = \frac{1}{s}\frac{1+\frac{e^\frac{x-\mu}{2s}-e^\frac{-(x-\mu)}{2s}}{e^\frac{x-\mu}{2s}+e^\frac{-(x-\mu)}{2s}}}{\bigl(e^\frac{x-\mu}{2s}+e^\frac{-(x-\mu)}{2s}\bigr)^2} = \frac{2e^\frac{x-\mu}{2s}}{s\bigl(e^\frac{x-\mu}{2s}+e^\frac{-(x-\mu)}{2s}\bigr)^3}
\end{align}

```python
def pdf(x,μ,s):
    return (2/s*np.exp((x-μ)/(2*s)))/((np.exp((x-μ)/(2*s)))+np.exp(-(x-μ)/(2*s)))**3
x = np.linspace(-15,30,10000)
pdfs = pdf(x,μ,s)
```

To draw the PDF ontop of the histogram we first need to normalize the histogram. This is done with the "denisty = true " option.

```python
plt.hist(xs,bins=80,density = True)
plt.plot(x,pdfs,label="Analytical PDF")
plt.xlabel("x")
plt.ylabel("Probablity density")
plt.vlines(np.mean(xs),0,1,color="red",label=f"Mean m = {round(np.mean(xs),2)}")
plt.xlim(-15,25)
plt.ylim(0,0.1)
plt.title("Inverse Transform Sampling")
plt.legend()
plt.show()
```
